##### File Operations (Upload Image, Delete Image, Display Image Side-by-Side, and Save Image)

```python
# Function to upload an image
def upload_image():
    global original_img, processed_img, img_canvas1, img_canvas2, img_path
    file_path = filedialog.askopenfilename()
    if file_path:
        original_img = cv2.imread(file_path) 
        img_path = file_path
        processed_img = original_img.copy()
        display_image(original_img, img_canvas1)
        display_image(processed_img, img_canvas2)

# Function to delete the image
def delete_image():
    global original_img, processed_img, img_canvas1, img_canvas2, img_path
    img_canvas1.delete("all")
    img_canvas2.delete("all")
    
# Function to resize and display an image side by side in a canvas
def display_image(img, canvas):
    canvas_width = canvas.winfo_width()
    canvas_height = canvas.winfo_height()
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_pil = Image.fromarray(img_rgb)
    img_pil.thumbnail((canvas_width, canvas_height), Image.LANCZOS)
    img_tk = ImageTk.PhotoImage(image=img_pil)
    canvas.delete("all")
    canvas.create_image(0, 0, anchor='nw', image=img_tk)
    canvas.image = img_tk

# Function to save the processed image
def save_image():
    global processed_img
    file_path = filedialog.asksaveasfilename(defaultextension=".jpg")
    if file_path:
        cv2.imwrite(file_path, processed_img)
```

##### Drawing Shapes on Image

```python
line_thickness = 1

# Function to update the line thickness based on slider value
def update_thickness(value):
    global line_thickness
    line_thickness = int(value)

# Function to draw a line on the image
def draw_line(event):
    global ix, iy, drawing, processed_img, lines

    x, y = event.x, event.y
    if event.type == tk.EventType.ButtonPress:
        drawing = True
        ix, iy = x, y
    elif event.type == tk.EventType.Motion:
        if drawing:
            temp_img = processed_img.copy()
            cv2.line(temp_img, (ix, iy), (x, y), (0, 200, 255), line_thickness)
            display_image(temp_img, img_canvas2)
    elif event.type == tk.EventType.ButtonRelease:
        drawing = False
        cv2.line(processed_img, (ix, iy), (x, y), (0, 200, 255), line_thickness)
        lines.append(((ix, iy), (x, y), line_thickness))
        display_image(processed_img, img_canvas2)

# Function to draw a circle
def draw_circle(event):
    global ix, iy, drawing, processed_img

    x, y = event.x, event.y
    if event.type == tk.EventType.ButtonPress:
        drawing = True
        ix, iy = x, y
    elif event.type == tk.EventType.Motion:
        if drawing:
            temp_img = processed_img.copy()
            radius = int(((x - ix) ** 2 + (y - iy) ** 2) ** 0.5)
            cv2.circle(temp_img, (ix, iy), radius, (0, 200, 255), line_thickness)
            display_image(temp_img, img_canvas2)
    elif event.type == tk.EventType.ButtonRelease:
        drawing = False
        radius = int(((x - ix) ** 2 + (y - iy) ** 2) ** 0.5)
        cv2.circle(processed_img, (ix, iy), radius, (0, 200, 255), line_thickness)
        display_image(processed_img, img_canvas2)

# Function to draw a rectangle 
def draw_rectangle(event):
    global ix, iy, drawing, processed_img

    x, y = event.x, event.y
    if event.type == tk.EventType.ButtonPress:
        drawing = True
        ix, iy = x, y
    elif event.type == tk.EventType.Motion:
        if drawing:
            temp_img = processed_img.copy()
            cv2.rectangle(temp_img, (ix, iy), (x, y), (0, 200, 255), line_thickness)
            display_image(temp_img, img_canvas2)
    elif event.type == tk.EventType.ButtonRelease:
        drawing = False
        cv2.rectangle(processed_img, (ix, iy), (x, y), (0, 200, 255), line_thickness)
        display_image(processed_img, img_canvas2)

# Function to draw a triangle
def draw_triangle(event):
    global ix, iy, drawing, processed_img

    x, y = event.x, event.y
    if event.type == tk.EventType.ButtonPress:
        drawing = True
        ix, iy = x, y
    elif event.type == tk.EventType.Motion:
        if drawing:
            temp_img = processed_img.copy()
            points = np.array([[ix, iy], [x, y], [(ix + x) // 2, iy - (y - iy)]], np.int32)
            cv2.polylines(temp_img, [points], isClosed=True, color=(0, 200, 255), thickness=line_thickness)
            display_image(temp_img, img_canvas2)
    elif event.type == tk.EventType.ButtonRelease:
        drawing = False
        points = np.array([[ix, iy], [x, y], [(ix + x) // 2, iy - (y - iy)]], np.int32)
        cv2.polylines(processed_img, [points], isClosed=True, color=(0, 200, 255), thickness=line_thickness)
        display_image(processed_img, img_canvas2)

# Function to draw a square
def draw_square(event):
    global ix, iy, drawing, processed_img

    x, y = event.x, event.y
    if event.type == tk.EventType.ButtonPress:
        drawing = True
        ix, iy = x, y
    elif event.type == tk.EventType.Motion:
        if drawing:
            temp_img = processed_img.copy()
            side_length = min(abs(x - ix), abs(y - iy))
            cv2.rectangle(temp_img, (ix, iy), (ix + side_length, iy + side_length), (0, 200, 255), line_thickness)
            display_image(temp_img, img_canvas2)
    elif event.type == tk.EventType.ButtonRelease:
        drawing = False
        side_length = min(abs(x - ix), abs(y - iy))
        cv2.rectangle(processed_img, (ix, iy), (ix + side_length, iy + side_length), (0, 200, 255), line_thickness)
        display_image(processed_img, img_canvas2)
```

### Main UI Components

```python
# Custom Button Design
class CustomButton1(tk.Button):
    def __init__(self, parent, text, command):
        super().__init__(parent, text=text, command=command)
        self.configure(
            background=color2,
            foreground=white,
            padx=15,
            activebackground=color3,
            activeforeground=white,
            width=10,
            height=1,
            relief='raised',
            border=5, 
            cursor='hand2',
            font=('Arial', 16, 'bold')
        )

class CustomButton2(tk.Button):
    def __init__(self, parent, text, command):
        super().__init__(parent, text=text, command=command)
        self.configure(
            background=gray,
            foreground=white,
            padx=15,
            activebackground=gray,
            activeforeground=color2,
            width=10,
            height=1,
            border=4,
            relief='groove',
            cursor='hand2',
            font=('Arial', 13, 'bold')
        ) 

#colors used to create ui
gray = '#676767'
dGray = '#4C4C4C'
lGray = '#989898'
black = 'BLACK'
white = 'WHITE'
color1 = '#020f12'
color2 = '#0387E7'
color3 = '#05C3FF'

# Main application window
root = tk.Tk()
root.title("PicMagic")
root.state('zoomed') 
root.config(bg= gray)

# Style for the notebook and tabs
style = ttk.Style()
style.theme_use('default')

style.configure('TNotebook', background=gray, borderwidth=0)
style.configure('TNotebook.Tab', background=color2, foreground=white, font=('Arial', 11, 'bold'),anchor='center', justify='center', padding=[4, 5], bd=0,width=10)
style.map('TNotebook.Tab', background=[('selected', color3)])

# Notebook for the sidebar
sidebar = ttk.Notebook(root)

# Tabs for each tool set in the sidebar
tab1 = tk.Frame(sidebar, bg=dGray, bd=0)
tab2 = tk.Frame(sidebar, bg=dGray, bd=0)
tab3 = tk.Frame(sidebar, bg=dGray, bd=0)

sidebar.add(tab1, text="Basics")
sidebar.add(tab2, text="Advanced")
sidebar.add(tab3, text="AI Tools")

sidebar.pack(side=tk.LEFT, fill='y')

# Display the tools in each tab
label2 = tk.Label(tab2, text='Advanced Editing Tools')
label2.pack()
label3 = tk.Label(tab3, text='AI Editing Tools')
label3.pack()

# Frame for upload and delete buttons
btn_frame1 = tk.Frame(root,bg = gray)
btn_frame1.pack(side=tk.TOP,pady=10)

# Frame for shape drawing buttons
btn_frame2 = tk.Frame(tab1,bg = lGray)
btn_frame2.pack(side=tk.TOP,pady=20)

# Upload image button
upload_btn = CustomButton1(btn_frame1, text="Upload image", command=upload_image)
upload_btn.grid(row=0, column=0,padx=5)

# Delete image button
delete_btn = CustomButton1(btn_frame1, text="Delete image", command=delete_image)
delete_btn.grid(row=0, column=1,padx=5)

# Save image button
save_btn = CustomButton1(root, text="Save image", command=save_image)
save_btn.pack(side=tk.BOTTOM, pady=15, padx=(0, 15), anchor="se")

#line thickness slider
thickness_slider = tk.Scale(btn_frame2, from_=1, to=100, orient='horizontal', length=200, label="Line Thickness", command=update_thickness)
thickness_slider.set(line_thickness)  # Set the initial thickness
thickness_slider.pack(pady=10)

# Draw Line button
draw_line_btn = CustomButton2(btn_frame2, text="Draw Line", command=drawing_line)
draw_line_btn.pack(pady=5)

# Draw Circle button
draw_circle_btn = CustomButton2(btn_frame2, text="Draw Circle", command=drawing_circle)  
draw_circle_btn.pack(pady=5)

# Draw Rectangle button
draw_rectangle_btn = CustomButton2(btn_frame2, text="Draw Rectangle", command=drawing_rectangle) 
draw_rectangle_btn.pack(pady=5)

# Draw Triangle button
draw_triangle_btn = CustomButton2(btn_frame2, text="Draw Triangle", command=drawing_triangle) 
draw_triangle_btn.pack(pady=5)

# Draw Square button
draw_square_btn = CustomButton2(btn_frame2, text="Draw Square", command=drawing_square) 
draw_square_btn.pack(pady=5)

# Display area for images
img_frame = tk.Frame(root)
img_frame.pack(side=tk.TOP, padx=10)

img_canvas1 = tk.Canvas(img_frame,width= 610,height=630,bg= white,bd=0,highlightthickness=3, highlightbackground=gray)
img_canvas1.pack(side=tk.LEFT)

img_canvas2 = tk.Canvas(img_frame,width=610,height=630,bg= white,bd=0,highlightthickness=3, highlightbackground=gray)
img_canvas2.pack(side=tk.RIGHT)

# Global variables to hold the images
original_img = None
processed_img = None
img_path = None

root.mainloop()
```
