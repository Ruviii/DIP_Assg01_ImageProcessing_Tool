##### Main UI Components

```python
import tkinter as tk
from tkinter import filedialog, ttk
from PIL import Image, ImageTk, ImageEnhance

# Stack to hold history for undo/redo functionality
undo_stack = []
redo_stack = []

# Function to rotate the image
def rotate_image():
    global processed_img
    if processed_img:
        save_undo_state()
        processed_img = processed_img.rotate(45)
        display_image(processed_img, img_canvas2)

# Function to crop the image
def crop_image():
    global processed_img
    if processed_img:
        save_undo_state()
        processed_img = processed_img.crop((100, 100, 400, 400))
        display_image(processed_img, img_canvas2)

# Function to flip the image
def flip_image():
    global processed_img
    if processed_img:
        save_undo_state()
        processed_img = processed_img.transpose(Image.FLIP_LEFT_RIGHT)
        display_image(processed_img, img_canvas2)

# Function to change the brightness of the image
def change_brightness():
    global processed_img
    if processed_img:
        save_undo_state()
        enhancer = ImageEnhance.Brightness(processed_img)
        processed_img = enhancer.enhance(1.5)  # Increase brightness by 50%
        display_image(processed_img, img_canvas2)

# Function to save current image state for undo functionality
def save_undo_state():
    global undo_stack, processed_img
    if processed_img:
        undo_stack.append(processed_img.copy())
        redo_stack.clear()  # Clear redo stack when new action is performed

# Function to undo the last action
def undo_action():
    global undo_stack, redo_stack, processed_img
    if undo_stack:
        redo_stack.append(processed_img.copy())
        processed_img = undo_stack.pop()
        display_image(processed_img, img_canvas2)

# Function to redo the last undone action
def redo_action():
    global undo_stack, redo_stack, processed_img
    if redo_stack:
        undo_stack.append(processed_img.copy())
        processed_img = redo_stack.pop()
        display_image(processed_img, img_canvas2)

# Custom Button Design
class CustomButton(tk.Button):
    def __init__(self, parent, text, command):
        super().__init__(parent, text=text, command=command)
        self.configure(
            background=color2,
            foreground=white,
            padx=15,
            activebackground=color3,
            activeforeground=white,
            width=10,
            height=2,
            border=0,
            relief='flat', 
            cursor='hand2',
            font=('Arial', 16, 'bold')
        )

# Function to upload an image        
def upload_image():
    global original_img, processed_img, img_canvas1, img_canvas2, img_path
    file_path = filedialog.askopenfilename()
    if file_path:
        original_img = Image.open(file_path)
        img_path = file_path
        display_image(original_img, img_canvas1)
        processed_img = original_img.copy()
        display_image(processed_img, img_canvas2)
        undo_stack.clear()  # Clear the undo/redo stack when new image is uploaded
        redo_stack.clear()

# Function to delete the image 
def delete_image():
    global original_img, processed_img, img_canvas1, img_canvas2, img_path
    img_canvas1.delete("all")
    img_canvas2.delete("all")

# Function to display an image in a canvas
def display_image(img, canvas):
    img_tk = ImageTk.PhotoImage(img)
    canvas.create_image(0, 0, anchor='nw', image=img_tk)
    canvas.image = img_tk

# Function to save the processed image
def save_image():
    global processed_img
    file_path = filedialog.asksaveasfilename(defaultextension=".jpg")
    if file_path:
        processed_img.save(file_path) 

# Colors
gray = '#676767'
dGray = '#4C4C4C'
black = 'BLACK'
white = 'WHITE'
color1 = '#020f12'
color2 = '#0387E7'
color3 = '#05C3FF'

# Main application window
root = tk.Tk()
root.title("PicMagic")
root.state('zoomed') 
root.config(bg=gray)

# Notebook for the sidebar
sidebar = ttk.Notebook(root, width=300, height=600)

# Tabs for each tool set in the sidebar
tab1 = tk.Frame(sidebar)
tab2 = tk.Frame(sidebar)
tab3 = tk.Frame(sidebar)

sidebar.add(tab1, text="Basics")
sidebar.add(tab2, text="Enhancements")
sidebar.add(tab3, text="AI Tools")
sidebar.pack(side=tk.LEFT, fill='y')

# Display the tools in each tab
label1 = tk.Label(tab1, text='Basic Editing Tools', width=50, height=2).pack()

# Rotate, Crop, and Flip buttons under "Basics"
rotate_btn = CustomButton(tab1, text="Rotate", command=rotate_image)
rotate_btn.pack(pady=10)

crop_btn = CustomButton(tab1, text="Crop", command=crop_image)
crop_btn.pack(pady=10)

flip_btn = CustomButton(tab1, text="Flip", command=flip_image)
flip_btn.pack(pady=10)

# Undo, Redo, and Change Brightness buttons
undo_btn = CustomButton(tab1, text="Undo", command=undo_action)
undo_btn.pack(pady=10)

redo_btn = CustomButton(tab1, text="Redo", command=redo_action)
redo_btn.pack(pady=10)

brightness_btn = CustomButton(tab1, text="Brightness", command=change_brightness)
brightness_btn.pack(pady=10)

# Frame for upload and delete buttons
btn_frame1 = tk.Frame(root, bg=gray)
btn_frame1.pack(side=tk.TOP, anchor="w", pady=20)

# Upload image button
upload_btn = CustomButton(btn_frame1, text="Upload image", command=upload_image)
upload_btn.grid(row=0, column=0, padx=20)

# Delete image button
delete_btn = CustomButton(btn_frame1, text="Delete image", command=delete_image)
delete_btn.grid(row=0, column=1)

# Save image button
save_btn = CustomButton(root, text="Save image", command=save_image)
save_btn.pack(side=tk.BOTTOM, pady=30, padx=(0, 20), anchor="se")

# Display area for images
img_frame = tk.Frame(root)
img_frame.pack(side=tk.TOP, padx=20)

img_canvas1 = tk.Canvas(img_frame, width=600, height=600, bg=white, bd=0, highlightthickness=3, highlightbackground=dGray)
img_canvas1.pack(side=tk.LEFT)

img_canvas2 = tk.Canvas(img_frame, width=600, height=600, bg=white, bd=0, highlightthickness=3, highlightbackground=dGray)
img_canvas2.pack(side=tk.RIGHT)

# Global variables to hold the images
original_img = None
processed_img = None
img_path = None

# Start the main event loop
root.mainloop()
```
